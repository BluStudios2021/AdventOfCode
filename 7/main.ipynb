```python
with open("./input.txt", "r") as f:
    lines = f.read().splitlines()
```

```python
def getCurrentDict(tree, path):
    current = tree
    for a in path[:-1]:
        current = current["/" + a]
    return current

def loopFilesystem(tree, parent):
    sizes = []
    count = 0
    for key, value in tree.items():
        if key[0] == "/":
            childSizes = loopFilesystem(value, key[1:])
            for size in childSizes:
                sizes.append(size)
            count += childSizes[-1]
            if key in ["/qzgjp", "/tzssf"]:
                print(key[1:])
                print(childSizes)
        else:
            count += int(value)
    sizes.append(count)
    return sizes
```

```python
# create filesystem
import json

path = []
tree = {}
current = {}

for line in lines:
    # command
    if line[0] == "$":
        command = line[2:4]
        # change directory
        if command == "cd":
            val = line[5:]
            # go back
            if val == "..":
                path.pop()
                current = getCurrentDict(tree, path)
            # set directory
            else:
                path.append(val)
                current = getCurrentDict(tree, path)
                if not "/" + val in current:
                    current["/" + val] = {}

        elif command == "ls":
            pass
    elif line[0:3] == "dir":
        """
        val = line[4:]
        current = getCurrentDict(tree, path)
        if not val in current:
            current[val] = {}
        """
        pass
    else:
        size, name = line.split(" ")
        current[name] = size

pretty = json.dumps(tree, indent=2)
print(pretty)
```

```python
sizes = loopFilesystem(tree["//"], "/")

print(sizes)
```

```python
count = 0

for value in sizes:
    if value <= 100000:
        count += value

print(count)
```

```python
def countFolders(tree):
    count = 0
    for key, value in tree.items():
        if key[0] == "/":
            count += countFolders(value) + 1

    return count

print(countFolders(tree))
print(len(sizes))
```
