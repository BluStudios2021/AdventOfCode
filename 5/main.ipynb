```python
moves = []

with open("./input.txt", "r") as f:
    lines = f.read().splitlines()

for line in lines:
    strings = line.split(" ")
    move = [int(strings[1]), int(strings[3]), int(strings[5])]
    moves.append(move)

print(moves)
```

```python
"""
    [C]         [Q]         [V]    
    [D]         [D] [S]     [M] [Z]
    [G]     [P] [W] [M]     [C] [G]
    [F]     [Z] [C] [D] [P] [S] [W]
[P] [L]     [C] [V] [W] [W] [H] [L]
[G] [B] [V] [R] [L] [N] [G] [P] [F]
[R] [T] [S] [S] [S] [T] [D] [L] [P]
[N] [J] [M] [L] [P] [C] [H] [Z] [R]
 1   2   3   4   5   6   7   8   9 
"""
```

```python
# part one
stacks = [
["N", "R", "G", "P"],
["J", "T", "B", "L", "F", "G", "D", "C"],
["M", "S", "V"],
["L", "S", "R", "C", "Z", "P"],
["P", "S", "L", "V", "C", "W", "D", "Q"],
["C", "T", "N", "W", "D", "M", "S"],
["H", "D", "G", "W", "P"],
["Z", "L", "P", "H", "S", "C", "M", "V"],
["R", "P", "F", "L", "W", "G", "Z"]
]

for move in moves:
    for i in range(move[0]):
        a = stacks[move[1] - 1].pop()
        stacks[move[2] - 1].append(a)

output = ""

for stack in stacks:
    output += stack[-1]

print(output)
```

```python
# part two
stacks = [
["N", "R", "G", "P"],
["J", "T", "B", "L", "F", "G", "D", "C"],
["M", "S", "V"],
["L", "S", "R", "C", "Z", "P"],
["P", "S", "L", "V", "C", "W", "D", "Q"],
["C", "T", "N", "W", "D", "M", "S"],
["H", "D", "G", "W", "P"],
["Z", "L", "P", "H", "S", "C", "M", "V"],
["R", "P", "F", "L", "W", "G", "Z"]
]

for move in moves:
    for i in range(move[0]):
        stack = stacks[move[1] - 1]
        a = stack.pop(len(stack) - (move[0] - i))
        stacks[move[2] - 1].append(a)



output = ""

for stack in stacks:
    output += stack[-1]

print(output)
```
